# 按季度进行划分

+ 第1季度：1-3月
+ 第2季度：4-6月
+ 第3季度：7-9月
+ 第4季度：10-12月

```python
import pandas as pd

# 季度月份
startMonth = 1
endMonth = 3

# 标签文件
labels = pd.read_csv('labels.csv')
labels['date'] = pd.to_datetime(labels['date'])

# 测试集
test = labels[('2020-01-01' <=labels['date']) & (labels['date'] <= '2025-01-01')]
test.to_csv('test.csv',index=False)

# 训练集+验证集
trainVal = labels[labels['date'] < '2020-01-01']
train = trainVal[~((startMonth <= trainVal['date'].dt.month) & (trainVal['date'].dt.month <= endMonth))]
val = trainVal[(startMonth <= trainVal['date'].dt.month) & (trainVal['date'].dt.month <= endMonth)]
train.to_csv('train.csv',index=False)
val.to_csv('val.csv',index=False)
```
